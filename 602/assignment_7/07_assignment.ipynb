After performing all those changes, we can ask the question "Per year, what cause has the highest amount of deaths" and we can also do that for death_rate! 

```python
## Data Wrangling

import pandas as pd

cat_cols = ["year", "leading_cause", "sex", "race_ethnicity"]
num_cols = ["deaths", "death_rate", "age_adjusted_death_rate"]

# Rename columns
death_df = death_df.rename(columns={
    "leading_cause": "cause",
    "race_ethnicity": "race",
    "age_adjusted_death_rate": "age_adj_death_rate"
})

# Update column lists after renaming
cat_cols = ["year", "cause", "sex", "race"]
num_cols = ["deaths", "death_rate", "age_adj_death_rate"]

# Standardize sex values
death_df["sex"] = (
    death_df["sex"]
    .str.strip()
    .replace({
        "M": "Male",
        "F": "Female"
    })
)

# Create new column: difference between age-adjusted and crude death rate
death_df["adj_rate"] = (
    death_df["age_adj_death_rate"] - death_df["death_rate"]
)

# Drop rows with missing rate values
death_df = death_df.dropna(
    subset=["death_rate", "age_adj_death_rate"]
)

# Optional: drop race column
# death_df = death_df.drop(columns=["race"])

# Sort by year and deaths
death_df = death_df.sort_values(
    ["year", "deaths"],
    ascending=[True, False]
)

# Filter: only male records
male_df = death_df[death_df["sex"] == "Male"]

# Check if cause is numeric
cause_is_numeric = pd.api.types.is_numeric_dtype(death_df["cause"])
print("Is cause numeric?", cause_is_numeric)

# Group and summary statistics by cause
cause_summary = death_df.groupby("cause")["deaths"].describe()

# Group by year and cause, then sort within year
year_cause_deaths = (
    death_df
    .groupby(["year", "cause"], as_index=False)["deaths"]
    .sum()
    .sort_values(["year", "deaths"], ascending=[True, False])
)

display(death_df.head())
display(cause_summary)
display(year_cause_deaths.head(20))
```

```python
# Question

# Per year, cause with the highest total deaths
top_deaths_per_year = (
    death_df
    .groupby(["year", "cause"], as_index=False)["deaths"]
    .sum()
    .loc[lambda df: df.groupby("year")["deaths"].idxmax()]
    .sort_values("year")
)

display(top_deaths_per_year)

# Per year, cause with the highest average death rate
top_death_rate_per_year = (
    death_df
    .groupby(["year", "cause"], as_index=False)["death_rate"]
    .mean()
    .loc[lambda df: df.groupby("year")["death_rate"].idxmax()]
    .sort_values("year")
)

display(top_death_rate_per_year)
```
