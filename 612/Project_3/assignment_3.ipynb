The title **Recommenders: The Cause of All Society’s Problems?** is a tongue-in-cheek joke. It mocks the idea that these data systems are often blamed for social issues, when they are really tools that can be used appropriately or inappropriately.

## In what ways do you think Recommender Systems reinforce human bias?

Recommender systems try to create signals from user data, which often amplifies what a specific user or group of users already wants. That is what these systems were created for, whether it is recommending music on Spotify, recommending job candidates on Indeed, or recommending political commentary on X.

The recent recommender techniques we learned, like content-based filtering, item-item filtering, and SVD, all rely on previous behavior and ratings in order to suggest possible recommendations. This will, of course, reinforce bias. While you may get some different recommendations through item-item or user-user filtering, the overall bias is still being perpetuated because the system is learning from past behavior.


## Do you think recommender systems reinforce or help to prevent unethical targeting or customer segmentation?

Reinforce.

Similar to the previous answer, recommenders are just mathematical models that learn from user behavior. There is nothing unethical on the side of the recommenders. The main blame should be placed on users, devs, and society. If a user shows biased preferences, the recommender will push those preferences. They may amplify customer-segmented groups, but that is a normal everyday part of society.

From the readings, I think there is too much focus on the potential destructive nature of these systems. It leads people to believe that engineers should act as moral arbiters by inserting their own notions of fairness into the models.

In the [Evan Estola](https://www.youtube.com/watch?v=MqoRzNhrTnQ) video, he argues that ethics in machine learning is not just an abstract concept, but a technical challenge that requires intentional design choices to ensure fairness, accuracy, and transparency. He also mentions that engineers should have “social responsibility.”

I disagree. I think it causes issues, like we saw in old Twitter. It is just better to code without intentional bias, and if the outcomes based on user input are not desirable, that is something society has to figure out.
