## Music Recommendations at Scale w/ Spark - Spotify

I guess the most interesting point is the math, Alternating Least Squares (ALS), which is still a popular method to solve for matrix factorization in music recommenders. It's pretty brute force, because you're factorizing a (usually very sparse) user-item interaction matrix into a User Matrix (X) and Item Matrix (Y) by using ALS, which iterates between the two matrices and performs a calculation until it reaches convergence.

Kinda magical, kinda brute. 

Another cool part is the idea of vectors allow you to scale easier, because you can perform operations on millions of vectors at once, since there's no dependence on one vector to another. However when these calculations get so massive with the amount of users and songs, you run into issues with performance. So, the half-gridify method they discussed is a way to arrange data so there's not so much reading, writing, and swapping between server clusters. This is still an issue today, especially with AI and hyperscalers, even with new techniques and quicker hardware, people always push the boundaries of what's possible.
