# Mitigating the Harm of Recommender Systems

Recommender systems can reduce radicalization and discrimination by focusing on user well-being, fairness, and content diversity rather than maximizing clicks or watch time. 

Platforms should limit repeated recommendations of increasingly extreme content, provide users with more control over their feeds, and regularly audit outcomes across demographic groups — today, extreme content can be flagged automatically via an AI review, similar to how X.com does it. They can also model and correct social-influence bias so that popular ratings or opinions do not distort users’ genuine preferences. Together, these measures would weaken harmful feedback loops while preserving useful personalization.
