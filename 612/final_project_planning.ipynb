# Final Project Planning Document

For my final project, I want to build on my previous MovieLens recommender by adapting the system to Spark and using a larger version of the data. My earlier project required heavy filtering because of memory limits, so this project will focus on whether Spark makes it practical to use more of the MovieLens dataset locally.

The main system will use MovieLens ratings to generate candidate recommendations. I plan to use Spark in local mode, which means the project will run on my own machine rather than a full cluster. Even so, Spark should let me use partitioned data and parallel processing across local CPU cores. Since my machine has a Ryzen 9 9900X CPU and an RTX 4080 GPU, I should be able to test larger versions of the dataset locally. If time allows, I may also test a small PyTorch model as a GPU-based extension.

The unique part of the project will be using the MovieLens tag genome data as part of the ranking process. In my previous project, I used genres mostly as a basic diversity filter. This time, I want to use tag genome scores more directly. The tag genome gives numeric relevance scores for many movie tags, so each movie can be represented as a more detailed feature vector instead of only broad genre labels.

The recommender will first generate candidate movies from the rating data. After that, the tag genome data will be used to rank those candidates more intelligently. For each user, I will create a taste profile from the tag genome vectors of movies they rated highly. Candidate movies will then be ranked using their rating-based recommendation score, their similarity to the user’s tag profile, and a redundancy penalty that avoids recommending too many movies with very similar tag profiles.

I plan to evaluate the Spark version against my previous implementation by looking at recommendation quality, runtime, data size, and implementation complexity. I will use metrics such as RMSE, Precision@K, and Recall@K where practical. I will also discuss when Spark becomes worth the added complexity for this kind of recommender system.
